```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

## Lê dados

```python
df=pd.read_csv('../data/imoveis_treino.csv', sep=";", index_col="Índice")
df_test=pd.read_csv('../data/imoveis_teste.csv', sep=";", index_col="Índice")

dic_rename = {"N° de Quartos": "n_quartos", "N de Quartos": "n_quartos"}

LE = LabelEncoder()
LE_tipo = LabelEncoder()

df = df.rename(columns=dic_rename)
df.Bairro = pd.Categorical(df.Bairro)
df['bairro_code'] = LE.fit_transform(df['Bairro'])
df['tipo_code'] = LE_tipo.fit_transform(df['Tipo'])


df_test = df_test.rename(columns=dic_rename)
df_test.Bairro = pd.Categorical(df_test.Bairro)
df_test['bairro_code'] = LE.fit_transform(df_test['Bairro'])
df_test['tipo_code'] = LE_tipo.fit_transform(df_test['Tipo'])

df.head()
```

## Análise exploratória

```python
print("valores nulos: ")
print(df.isna().sum())
print("--------------------------------------")
print("informacoes gerais: ")
print(df.info())
print("--------------------------------------")
print(df["Bairro"].value_counts())
print("--------------------------------------")
print(df["Tipo"].value_counts())
print(df["Vagas de Garagem"].value_counts())
```

## Divide o dataset

```python
X = df[["Tamanho", "n_quartos", "Vagas de Garagem", "bairro_code", "tipo_code"]]
y = df[["Valor de Venda"]]

X_test = df_test[["Tamanho", "n_quartos", "Vagas de Garagem", "bairro_code", "tipo_code"]]
y_test = df_test[["Valor de Venda"]]
X_test.head()
```

```python
lm=LinearRegression()
lm.fit(X,y)

print(lm.intercept_,lm.coef_)
```

```python
y_pred = lm.predict(X_test)
y_pred
```

```python
print("Erro quadrático Médio: %.2f" % mean_squared_error(y_test, y_pred))
print("R2: %.4f" % r2_score(y_test, y_pred))
```
