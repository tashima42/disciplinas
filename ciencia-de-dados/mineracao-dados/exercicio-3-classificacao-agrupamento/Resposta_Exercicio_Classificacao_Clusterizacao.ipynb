```python
from sklearn import tree
import pandas as pd 


dfTitanic = pd.read_csv('titanic.csv')

dfTitanic['Sex'] = pd.Categorical(dfTitanic['Sex'])
dfTitanic['Sex_coded'] = dfTitanic['Sex'].cat.codes

#Função que facilita a validação cruzada (cross-validation)
from sklearn.model_selection import cross_val_score

#Não estamos dividindo o dataset em apenas dois como antes.
X = dfTitanic[["Pclass","Sex_coded","Age","Siblings/Spouses Aboard","Parents/Children Aboard","Fare"]]
y = dfTitanic['Survived']

tree_model = tree.DecisionTreeClassifier()

#aqui a estratégia de cross-valiadtion with k-fold=10 será usada.
scores = cross_val_score(tree_model, X, y, cv=10)

#Visão do desempenho baseada em 10 divisões.
print(scores.mean())

dfTitanic
```

# 3

```python
from sklearn.neighbors import KNeighborsClassifier

#Não estamos dividindo o dataset em apenas dois como antes.
X = dfTitanic[["Pclass","Sex_coded","Age","Siblings/Spouses Aboard","Parents/Children Aboard","Fare"]]
y = dfTitanic['Survived']

for k in range(2,8):
    print('k = '+str(k))
    
    neigh = KNeighborsClassifier(n_neighbors=k)

    scores = cross_val_score(neigh, X, y, cv=10)

    # resultado praticamente igual ao da árvore de decisão testada
    print(scores.mean())
```

K= 3 fornece o melhor resultado. Mas mesmo assim não é melhor do que os resultados obtidos com árvores. <br>
A partir de k=3, à medida que o k vai aumentando, observamos uma tendência de degradação dos resultados.
