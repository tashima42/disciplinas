### Propriedades e distribuição do número de caracteres e palavras em reviews

```python
df_reviews['char_len'] = df_reviews['text'].str.len()
df_reviews['word_len'] = df_reviews['text'].str.split().apply(len)

df_reviews[['char_len', 'word_len']].describe()
```

### Médias de reações e número de palavras agregadas por estabelecimento

```python
reviews_business_agg = (
    df_reviews.groupby('business_id')
      .agg(
          num_reviews=('text', 'count'),
          avg_useful=('useful', 'mean'),
          avg_cool=('cool', 'mean'),
          avg_funny=('funny', 'mean'),
          avg_word_len=('word_len', 'mean'),
      )
      .reset_index()
)

reviews_business_agg.sort_values('num_reviews', ascending=False).head(10)
```

### Comparação entre médias e número de reviews para estabelecimentos abertos e fechados
Não há uma diferença relevante entre as médias de de avaliações consideradas legais, divertidas, úteis, ou no número de palavras na review, porém estabelecimentos abertos tem, na média, aproximadamente, 66% mais reviews do que estabelecimentos fechados.

```python
bus_agg = df.merge(reviews_business_agg, on="business_id", how="left")

aberto_vs_fechado = (
    bus_agg.groupby("is_open")[["avg_useful", "avg_cool", "avg_funny", "avg_word_len", "num_reviews"]]
    .mean()
    .reset_index()
)
print("Aberto vs. Fechado")
aberto_vs_fechado
```

### Correlação entre as média de reações a avaliações e a quantidade de palavras nelas
Existe uma correlação moderada entre reviews consideradas legais e úteis, úteis e divertidas e uma correlação um pouco mais forte entre reviews consideradas divertidas e legais. Não existe correlação clara entre reações a reviews e a quantidade de palavras nela.

```python
corr_reacoes = bus_agg[["avg_useful", "avg_cool", "avg_funny", "avg_word_len"]].corr()
corr_reacoes
```

```python
sns.heatmap(corr_reacoes)
```

## Modelagem

```python
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
df_lat = df[["latitude", "longitude"]]
kms_por_radian = 6371.0088
dbscan_config = DBSCAN(eps=0.7 /kms_por_radian, min_samples=30, metric="haversine")
dbscan_results = dbscan_config.fit(np.radians(df_lat))


labels = dbscan_results.labels_
df["cluster"] = labels
unique_labels = list(set(labels.tolist()))

cmap = plt.get_cmap('hsv',len(unique_labels))

for label in unique_labels:
    plt.scatter(df_lat.loc[labels == label,'latitude'], df.loc[labels == label,'longitude'], color=cmap(label))

plt.scatter(df_lat.loc[labels == -1,'latitude'], df_lat.loc[labels == -1,'longitude'], c = 'black')   

plt.show()

cluster_stats = (
    df[df['cluster'] != -1]
    .groupby('cluster')['destaque']
    .agg(['count', 'mean'])
    .rename(columns={'mean': 'true_rate'})
    .reset_index()
)
cluster_stats
```
