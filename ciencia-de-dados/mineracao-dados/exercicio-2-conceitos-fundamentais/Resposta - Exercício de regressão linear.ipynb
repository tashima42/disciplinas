# 1
a) Faça um modelo de regressão linear simples utilizando a variável 'KmRodado' para prever a 'PrecoVenda'<br>
b) Calcule o R2 para o modelo criado.

```python
#1a
import pandas as pd
from sklearn import linear_model
import numpy as np

dfCarros = pd.read_csv('datasetCarros.csv')

y= dfCarros['PrecoVenda']
X_km = np.array(dfCarros['KmRodado']).reshape((-1, 1))


regr_km = linear_model.LinearRegression()

regr_km.fit(X_km, y)


print('Intercept:', regr_km.intercept_)
print('Slope:', regr_km.coef_) 
```

```python
#1b

r_sq = regr_km.score(X_km, y)
print('R2:', r_sq)
```

Podemos observar que o modelo criado acima não possui um bom desempenho.

# 2
a) Separe o dataset em teste (5%) e treino (95%). Use o método 'train_test_split' do sklearn; configure o parâmetro random_state=10. <br>

b) Treine um modelo de regressão linear múltipla no dataset de treino utilizando todas as variáveis (exceto 'Nome') para prever a 'PrecoVenda' e exiba os coeficientes do modelo.<br>

c) Avalie o modelo encontrado utilizando o dataset de teste. Calcule o R2 e MSE.

```python
#2a

from sklearn.model_selection import train_test_split

dfCarros = pd.read_csv('datasetCarros.csv')

# Codificando as colunas categoricas
dfCarros.replace({'TipoCombustivel':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
dfCarros.replace({'Trasmissao':{'Manual':0,'Automatic':1}},inplace=True)


y= dfCarros['PrecoVenda']
X = dfCarros.drop(['Nome','PrecoVenda'],axis=1)


X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.05, random_state=10)
```

```python
#2b

# Cria um objeto de regressão linear 
regr = linear_model.LinearRegression()

# Treina o modelo usando o conjunto de treino
regr.fit(X_treino, y_treino)

# Coeficientes
print("Coeficientes:", regr.coef_)

print(X.columns)
```

```python
#2c
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regr.predict(X_teste)

print("Mean squared error: %.2f" % mean_squared_error(y_teste, y_pred))

print("R2 : %.2f" % r2_score(y_teste, y_pred))
```
