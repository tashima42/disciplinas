# Questão 1

```python
from pyyoutube import Api

api = Api(api_key="SUA CREDENCIAL")
```

```python
r = api.search(location="-23.441284, -46.602701",
               location_radius="100mi",
               q="eleições",
               parts=["snippet"],
               count=50,
               published_after="2022-09-01T00:00:00Z",
               published_before="2022-10-30T00:00:00Z",
               safe_search="moderate",
               search_type="video")
```

```python

allVideos = {}
maior={'maiorVal':0,'idv':0}
for v in r.items:
    vid = v.to_dict()['id']['videoId']
    video = api.get_video_by_id(video_id=vid).to_dict()
    
    video_statistics = video['items'][0]['statistics']
    print(video_statistics)
    allVideos.update({vid:video})
    
    val = video_statistics['commentCount']
    if val != None and val > maior['maiorVal']:
        maior['maiorVal'] = val
        maior['idv'] = vid
        
    
    
```

# Questão 2

```python
import requests
from bs4 import BeautifulSoup


url="https://www.apontador.com.br/local/sp/campinas/faculdades_e_universidades/88Q2E372/unicamp.html"

##retorna o conteudo da pagina
req = requests.get(url)

##transforma o conteudo da pagina em um objeto BeautifulSoup
soup = BeautifulSoup(req.content,'html.parser')

conteudo = soup.find_all("div",{"class":"store-comments__text"})     
for review in conteudo:
    print(review.text)
```
