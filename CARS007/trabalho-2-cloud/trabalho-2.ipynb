```python
import pandas as pd
df = pd.read_csv("data/savedrecs.txt", sep="\t")
df
```

```python
from tabulate import tabulate
```

```python
df = df.reset_index(names="id")
```

```python
top_documents = pd.read_csv("data/top-10-documents.txt", sep="\t", usecols=["id", "document", "citations"])
sorted_documents = top_documents.sort_values(by="citations", ascending=False)
sorted_documents["id"] = sorted_documents["id"].apply(lambda i: i - 1)
sorted_documents.set_index("id")

df_merged = pd.merge(sorted_documents, df)
df_merged

df_merged.index += 1
print(tabulate(df_merged[["TI", "citations"]], headers = ["number", "title", "citations"], tablefmt = 'latex'))
```

```python
top_sources = pd.read_csv("data/top-10-sources.txt", sep="\t", usecols=["source", "documents", "citations"])
sorted_sources = top_sources.sort_values(by="citations", ascending=False)
sorted_sources = sorted_sources.reset_index()
sorted_sources.index += 1
print(tabulate(sorted_sources[["source", "citations", "documents"]], tablefmt="latex"))
```

## Most published countries

```python
top_countries = pd.read_csv("data/top-10-countries.txt", sep="\t", usecols=["country", "documents", "citations"])
sorted_countries = top_countries.sort_values(by="documents", ascending=False)
sorted_countries = sorted_countries.reset_index()
sorted_countries.index 
sorted_countries
```

```python
# Plot documents on the primary y-axis
fig, ax1 = plt.subplots(figsize=(16, 8))
color = 'tab:blue'
ax1.set_xlabel('País')
ax1.set_ylabel('Número de publicações', color=color)
ax1.bar(sorted_countries['country'], sorted_countries['documents'], color=color, label='Publicações')
ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='x', rotation=45)

# Create a second y-axis for citations
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Número de citações', color=color)
ax2.plot(sorted_countries['country'], sorted_countries['citations'], color=color, marker='o', linestyle='-', linewidth=2, label='Citações')
ax2.tick_params(axis='y', labelcolor=color)

# Title and legend
plt.title('Publicações e citações por país', pad=20)
fig.tight_layout()

plt.savefig("data/top-published-countries.eps", format='eps')
```

## Most published research institutions

```python
top_organizations = pd.read_csv("data/top-10-organizations.txt", sep="\t", usecols=["organization", "documents", "citations"])
sorted_organizations = top_organizations.sort_values(by="citations", ascending=False)
sorted_organizations = sorted_organizations.reset_index()
sorted_organizations
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))
color = 'tab:blue'
ax1.set_xlabel('Número de citações')
ax1.set_ylabel('Organização')
ax1.barh(sorted_organizations['organization'], sorted_organizations['citations'], label='Citações')
ax1.tick_params(axis='y')
ax1.tick_params(axis='x', rotation=90)

# Title and legend
plt.title('Citações por organização')
fig.tight_layout()

plt.savefig("data/top-cited-organizations.eps", format='eps')
```

## Publication trend over the years

```python
yearly_stats = df.groupby("PY").agg(
  total_published=("PY", "count"),
  total_citations=("TC", "sum")
).reset_index()

yearly_stats
```

```python
yearly_stats["total_published"].median()
```

```python
plt.figure(figsize=(16, 8))

for i, published in enumerate(yearly_stats["total_published"]):
    plt.text(yearly_stats["PY"][i], published + 5, str(published), ha='center')

# Add the number for each marker next to it
for i, citation in enumerate(yearly_stats["total_citations"]):
    height = 20
    if citation < 20:
        height += 40
    plt.text(yearly_stats["PY"][i], citation + height, str(citation), ha='center', color="purple")


plt.bar(yearly_stats["PY"], yearly_stats["total_published"], label="Total publicado")
plt.plot(yearly_stats["PY"], yearly_stats["total_citations"], marker='o', label="Total citado", color="purple")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.title("Evolução do número de publicações e citações por ano")
plt.legend()
plt.xticks(yearly_stats["PY"], rotation=45)
plt.tight_layout()
plt.savefig("data/number-literature-studies-published-by-year.eps", format='eps')
plt.show()
```
